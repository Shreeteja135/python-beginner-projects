```python
word = input("Enter your word : ")

keys = {
"abc" : 2,
"def" : 3,
"ghi" : 4,
"jkl" : 5,
"mno" : 6,
"pqrs": 7,
"tuv" : 8,
"wxyz": 9

}

for ch in word.lower():

    temp = ""

    for i in keys:
        if ch in i:
            temp = i
            break

    print(str(keys[temp])*(temp.find(ch)+1), end = "")
```
