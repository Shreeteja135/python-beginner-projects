```python
# Number guessing game
import random
unknown_number = random.randint(1,100)

count = 0
while True:
    guess = int(input("Enter your guess: ")) 
    count  += 1
    if guess == unknown_number:
        print(f" you guess the number {unknown_number} in {count} chance.....")
        break
    elif guess > unknown_number:
        print("Too high ..")
        print(f"try again ....")
    elif guess < unknown_number:
        print("Too low ..")
        print(f"try again ......")
   

```
